```python
import json
import os
from PIL import Image

# Update these paths according to your setup
json_labels_dir = '/Users/noramendoza/Desktop/GitHub/Synthesis-Project/yolov5/dataset/labels_json/train'
yolo_labels_dir = '/Users/noramendoza/Desktop/GitHub/Synthesis-Project/yolov5/dataset/labels/train'
images_dir = '/Users/noramendoza/Desktop/GitHub/Synthesis-Project/yolov5/dataset/images/train'

# Assuming your class names are mapped like this
class_mapping = {'question': 0, 'answer': 1, 'other': 2, 'header': 3}

# Create YOLO labels directory if it doesn't exist
os.makedirs(yolo_labels_dir, exist_ok=True)

def convert_box(size, box):
    dw = 1. / size[0]
    dh = 1. / size[1]
    x = (box[0] + box[2]) / 2.0 - 1
    y = (box[1] + box[3]) / 2.0 - 1
    w = box[2] - box[0]
    h = box[3] - box[1]
    x = x * dw
    w = w * dw
    y = y * dh
    h = h * dh
    return (x,y,w,h)

for json_file in os.listdir(json_labels_dir):
    if json_file.endswith(".json"):
        json_path = os.path.join(json_labels_dir, json_file)
        image_path = os.path.join(images_dir, json_file.replace('.json', '.png'))

        with Image.open(image_path) as img:
            width, height = img.size

        with open(json_path) as f:
            data = json.load(f)
            form_data = data.get('form', [])

            if not form_data:  # Skip if there are no annotations
                continue

            with open(os.path.join(yolo_labels_dir, json_file.replace('.json', '.txt')), 'w') as outfile:
                for item in form_data:
                    class_name = item.get('label')
                    if class_name in class_mapping:
                        class_id = class_mapping[class_name]
                        bbox = convert_box((width, height), item.get('box'))
                        outfile.write(f"{class_id} {' '.join(map(str, bbox))}\n")
        print(f"Processed {json_file}")
```
